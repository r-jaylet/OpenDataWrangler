```python
df_parcelles_vinci = pd.read_csv('../0_data/20230516_parcelles_vinci+immobilier.csv', sep=';')
```

```python
siren_parcelles = list(df_parcelles_vinci['N° SIREN (Propriétaire(s) parcelle)'].unique())
#siren_parcelles
```

```python
StockEtablissement = pd.read_csv('../0_data/base_sirene/StockEtablissement_utf8.csv', nrows=3)
StockEtablissement
```

```python
StockEtablissement = pd.read_csv('../0_data/base_sirene/StockEtablissement_utf8.csv', usecols=['siren', 'numeroVoieEtablissement', 
                                                                                               'indiceRepetitionEtablissement', 'typeVoieEtablissement',
                                                                                               'libelleVoieEtablissement', 'codePostalEtablissement', 'libelleCommuneEtablissement'])
```

```python
extract_StockEtablissement = StockEtablissement[StockEtablissement.siren.isin(siren_parcelles)]
```

```python
df_parcelles_vinci.columns
```

```python
extract_parcelles = df_parcelles_vinci[['N° SIREN (Propriétaire(s) parcelle)',
                                        'Nom Commune (Champ géographique)', 'N° de voirie (Adresse parcelle)', 
                                        'Nature voie (Adresse parcelle)', 'Nom voie (Adresse parcelle)',
                                        'Dénomination (Propriétaire(s) parcelle)']]
```

```python
extract_parcelles
```

```python
df = extract_parcelles.merge(extract_StockEtablissement, how='left', left_on='N° SIREN (Propriétaire(s) parcelle)',
                        right_on='siren')
```
