### Test via company search

```python
list_siren = [487501850, 432650273, 482980414]
```

```python
beneficiaires = pd.DataFrame()
entreprise = pd.DataFrame()
individu = pd.DataFrame()
```

```python
init = time.time()
for i in range(len(list_siren)):
    
    print('iteration = ', i)
    
    # change token
    if i % 100 == 0:
        token = get_token()
    siren = list_siren[i]
    
    df_beneficiaires, df_entreprise, df_individu = search_dirigeant(token, siren)  # vinci
    df_beneficiaires['siren'] = siren
    df_entreprise['siren'] = siren
    df_individu['siren'] = siren

    time.sleep(2)

    print('time searching : ', round(time.time() - init, 2), 's')
    print(len(df_beneficiaires), len(df_entreprise), len(df_individu))
    
    # add info on searched company
    entreprise = pd.concat([entreprise, df_entreprise])
    beneficiaires = pd.concat([beneficiaires, df_beneficiaires])
    individu = pd.concat([individu, df_individu])
    print(' ')
```

### Test via company updates

```python
res = pd.DataFrame(columns=['updatedAt', 'siren'])

init = time.time()
for i in range(2):
    
    print('iteration = ', i)

    endpoint = f"https://registre-national-entreprises.inpi.fr/api/companies/diff?page={i}&pageSize=100"
    headers = {"Authorization": "Bearer " + token}
    response = requests.get(endpoint, headers=headers).json()

    companies = [[r['company']['updatedAt'], r['company']['formality']['siren']] for r in response]
    df_companies = pd.DataFrame(companies, columns=['updatedAt', 'siren'])
    print('time searching : ', round(time.time() - init, 2), 's')
    
    res = pd.concat([res, df_companies])
    print(' ')
```

```python
res = pd.DataFrame(columns=['updatedAt', 'siren'])

d1 = 10
d2 = 11
init = time.time()
for i in range(5):
    
    print('iteration = ', i)

    d1+=1
    d2+=1
    
    endpoint = f"https://registre-national-entreprises.inpi.fr/api/companies/diff?pageSize=100&from=2020-05-{d1}&to=2020-05-{d2}"
    headers = {"Authorization": "Bearer " + token}
    response = requests.get(endpoint, headers=headers).json()

    companies = [[r['company']['updatedAt'], r['company']['formality']['siren']] for r in response]
    
    df_companies = pd.DataFrame(companies, columns=['updatedAt', 'siren'])
    print('time searching : ', round(time.time() - init, 2), 's')
    
    res = pd.concat([res, df_companies])
    print(' ')
```

### Get updates of 2023

```python
base = datetime.datetime.today()
date_list = [base - datetime.timedelta(days=x) for x in range(144)]
date_list = [date.strftime("%Y-%m-%d") for date in date_list]
date_list = list(reversed(date_list))
```

```python
res = pd.DataFrame(columns=['updatedAt', 'siren', 'date_searched'])
```

```python

init = time.time()
for i in range(40, len(date_list)-1):
    
    d1=date_list[i]
    d2=date_list[i+1]
    
    endpoint = f"https://registre-national-entreprises.inpi.fr/api/companies/diff?page=1&pageSize=100&zipCodes[]=92110&from={d1}&to={d2}"
    headers = {"Authorization": "Bearer " + token}
    response = requests.get(endpoint, headers=headers).json()
    companies = [[r['company']['updatedAt'], r['company']['formality']['siren'], d1] for r in response]
    df_companies = pd.DataFrame(companies, columns=['updatedAt', 'siren', 'date_searched'])
    
    print('time searching :', round(time.time() - init, 2), 's')

    print('# of companies found :', len(df_companies))
    
    res = pd.concat([res, df_companies])

    print(' ')
```
