### Process data

```python
SIREN_cols = pd.read_csv('../0_data/base_sirene/StockUniteLegale_utf8.csv', nrows=3)
```

```python
SIREN = pd.read_csv('../0_data/base_sirene/StockUniteLegale_utf8.csv', 
                    usecols=['siren', 'activitePrincipaleUniteLegale', 'denominationUniteLegale','etatAdministratifUniteLegale',
                             'unitePurgeeUniteLegale', 'caractereEmployeurUniteLegale', 'statutDiffusionUniteLegale',
                             'trancheEffectifsUniteLegale', 'prenom1UniteLegale', 'nomenclatureActivitePrincipaleUniteLegale',
                             'dateCreationUniteLegale', 'dateDernierTraitementUniteLegale'],
                    dtype = {'siren': str, 'trancheEffectifsUniteLegale': str})
```

```python
# remove old closed companies (les données concernant les entreprises cessées avant le 31/12/2002 ont été purgées)
SIREN = SIREN[SIREN.unitePurgeeUniteLegale.isna()]
```

```python
# only keep active companies
SIREN = SIREN[SIREN.etatAdministratifUniteLegale == 'A']
```

```python
# only keep companies that have shared all their data
SIREN = SIREN[SIREN.statutDiffusionUniteLegale == 'O']
```

```python
# Only keep Personnes Morales (remove personnes Physiques)
SIREN = SIREN[SIREN.prenom1UniteLegale.isna()]
```

```python
# remove unrelevent categories
SIREN['categorie'] = SIREN['activitePrincipaleUniteLegale'].apply(lambda x: x.split('.')[0])
#SIREN = SIREN[~SIREN.cat.isin(['01', '02', '03', '04', '05', '06', '07', '08', '99'])]

# only keep companies with more than 3 employees
#SIREN = SIREN[~SIREN.trancheEffectifsUniteLegale.isin(['NN', '00', '01'])]
```

```python
SIREN['dateDernierTraitementYEAR'] = SIREN['dateDernierTraitementUniteLegale'].dt.year
```

```python
SIREN.groupby('dateDernierTraitementYEAR').count()['siren'].plot(kind='bar')
```

### Analyse par tranche effectif

```python
SIREN.trancheEffectifsUniteLegale.isna().sum()
# More than half have null values
```

```python
SIREN_effectif = SIREN.dropna(subset ='trancheEffectifsUniteLegale')
```

```python
SIREN_effectif.groupby('trancheEffectifsUniteLegale').count()['siren'].plot(kind='barh')
```

- NN : Unité non employeuse (pas de salarié au cours de l'année de référence et pas d'effectif au 31/12)
- 00 : 0 salarié (n'ayant pas d'effectif au 31/12 mais ayant employé des salariés au cours de l'année de référence) 
- 01 : 1 ou 2 salariés 
- 02 : 3 à 5 salariés 
- 03 : 6 à 9 salariés 
- 11 : 10 à 19 salariés 
- 12 : 20 à 49 salariés 
- 21 : 50 à 99 salariés 
- 22 : 100 à 199 salariés 
- 31 : 200 à 249 salariés 
- 32 : 250 à 499 salariés 
- 41 : 500 à 999 salariés 
- 42 : 1 000 à 1 999 salariés 
- 51 : 2 000 à 4 999 salariés 
- 52 : 5 000 à 9 999 salariés 
- 53 : 10 000 salariés et plus

### Merge to SIRET to get siege loc info

```python
SIRET = pd.read_csv('../0_data/base_sirene/StockEtablissement_utf8.csv',
                    usecols=['siren', 'nic', 'siret', 'codePostalEtablissement','libelleCommuneEtablissement',
                             'etablissementSiege', 'etatAdministratifEtablissement'],
                    dtype = {'siren': str, 'siret': str, 'nic': str, 'codePostalEtablissement': str})
```

```python
SIRET_siege = SIRET[SIRET.etablissementSiege == True]
```

```python
SIRET_siege.isna().sum()
```

```python
len(SIRET_siege.siren.unique())  # check that siege well selected
```

```python
SIREN_loc = SIREN.merge(SIRET_siege, how='left', left_on = 'siren', right_on = 'siren')
```

```python
SIREN_loc.head(3)
```

```python
SIREN_loc = SIREN_loc.dropna(subset=['libelleCommuneEtablissement', 'codePostalEtablissement'])
SIREN_loc['departement'] = SIREN_loc['codePostalEtablissement'].apply(lambda x : x[:2])
```

```python
SIREN_loc = SIREN_loc[(SIREN_loc.departement != '00') & (SIREN_loc.departement != '[N')]
```

```python
figure(figsize=(20, 20))

SIREN_loc.groupby('departement').count()['siren'].plot(kind='barh')
```

### Focus Ile-de-France

```python
SIREN_IDF = SIREN_loc[SIREN_loc['departement'] == '75']
```

```python
SIREN_IDF.to_csv('../0_data/base_sirene/SIREN_IDF.csv')
```
