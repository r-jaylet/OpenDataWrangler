```python
import pandas as pd
import json
import seaborn
from matplotlib.pyplot import figure
from datetime import datetime
seaborn.set()
```

```python
pd.set_option('display.max_columns', None)
```

```python
SIREN = pd.read_csv('../0_data/base_sirene/StockUniteLegaleHistorique_utf8.csv', 
                    usecols=['siren', 'etatAdministratifUniteLegale', 'dateDebut', 'dateFin'],
                    dtype = {'siren': str})
```

### Analyse durée de vie entreprises

```python
SIREN['dateDebut'] = pd.to_datetime(SIREN['dateDebut'], format='mixed', dayfirst=True, errors='coerce')
SIREN['dateFin'] = pd.to_datetime(SIREN['dateFin'], format='mixed', dayfirst=True, errors='coerce')
SIREN = SIREN.sort_values('dateDebut', ascending=True)
```

```python
SIREN_last = SIREN.drop_duplicates('siren', keep='last')
SIREN_first = SIREN.drop_duplicates('siren', keep='first')

del SIREN
```

```python
SIREN_hist = SIREN_last.merge(SIREN_first[['siren', 'dateDebut']].rename({'dateDebut': 'dateStart'}, axis=1),
                              how='left', left_on='siren', right_on='siren')

del SIREN_first
del SIREN_last
```

```python
SIREN_hist.isna().sum()
```

```python
SIREN_hist_actifs = SIREN_hist[SIREN_hist.etatAdministratifUniteLegale == 'A']
SIREN_hist_fermes = SIREN_hist[SIREN_hist.etatAdministratifUniteLegale == 'C']
```

```python
del SIREN_hist
```

```python
SIREN_hist_actifs = SIREN_hist_actifs[(SIREN_hist_actifs.dateStart != '1900-01-01') & ~((SIREN_hist_actifs.dateStart.isna()))]
```

```python
SIREN_hist_actifs['delta'] = ((datetime.now() - SIREN_hist_actifs['dateStart']).dt.days / 365).fillna(0).astype('int', errors='ignore')
```

```python
SIREN_hist_fermes = SIREN_hist_fermes[(SIREN_hist_fermes.dateStart != '1900-01-01') & ~((SIREN_hist_fermes.dateStart.isna()))]
SIREN_hist_fermes = SIREN_hist_fermes[(SIREN_hist_fermes.dateDebut != '1900-01-01') & ~((SIREN_hist_fermes.dateDebut.isna()))]
```

```python
SIREN_hist_fermes['delta'] = ((SIREN_hist_fermes['dateDebut'] - SIREN_hist_fermes['dateStart']).dt.days / 365).fillna(0).astype('int', errors='ignore')
```

```python
figure(figsize=(16, 10))

SIREN_hist_actifs.loc[(SIREN_hist_actifs.delta >= 0) & (SIREN_hist_actifs.delta < 60)].groupby('delta').count()['siren'].plot(kind='bar', title='Distribution de la durée de vie des entreprises toujours en activitée en années (source : base historique SIREN)')
```

```python
figure(figsize=(16, 10))


SIREN_hist_fermes.loc[(SIREN_hist_fermes.delta >= 0) & (SIREN_hist_fermes.delta < 60)].groupby('delta').count()['siren'].plot(kind='bar', title='Distribution de la durée de vie des entreprises plus en activité en années (source : base historique SIREN)')
```
